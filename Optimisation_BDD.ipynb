```python
data = pd.read_csv('data_Us.csv')
data.head(3)
```

```python
def max_length(df):
    df_col_name = []  
    df_col_dtype = []  
    df_max_length = []  
    df_min_length = []
    df_header = ["Col Name", "Col Type", "Min" ,"Max"]
    for col in df.columns:
        df_col_name.append(col) 
        if df[col].dtypes =='object':
            df_col_dtype.append('Object_Type')
            max = df[col].apply(lambda x :len(x)).max()
            df_max_length.append(max)
            min = df[col].apply(lambda x :len(x)).min()
            df_min_length.append(min)
        elif df[col].dtypes=='int64':
            df_col_dtype.append('Entier')
            max = df[col].max()
            df_max_length.append(max)
            min = df[col].min()
            df_min_length.append(min)
        elif df[col].dtypes=='float64':
            df_col_dtype.append('Float')
            max = df[col].max()
            df_max_length.append(max)
            min = df[col].min()
            df_min_length.append(min)
        elif data[col].dtypes.kind == 'M':
            df_col_dtype.append('date/time (oldest most recent)') # Store the data type for the current col.
            max = data[col].max()  # Get the max  max number of current col (NaN removed).
            df_max_length.append(max)
            min = data[col].min()  # Get the max  max number of current col (NaN removed).
            df_min_length.append(min)
        elif data[col].dtypes =='bool':
            df_col_dtype.append('Boolean')
            max = data[col].max()  # Get the max  max number of current col (NaN removed).
            df_max_length.append(max)
            min = data[col].min()  # Get the max  max number of current col (NaN removed).
            df_min_length.append(min)
    result = list(zip(df_col_name, df_col_dtype, df_min_length, df_max_length)) # Merges the lists containing col names & max values.
    df_result = pd.DataFrame(result, columns = df_header)
    return df_result
    
```
